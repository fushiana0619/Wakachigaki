```python
from janome.tokenizer import Tokenizer
from gensim.models import word2vec
import re
```

```python
#バイナリデータとしてオープン
binarydata = open("gingatetsudono_yoru.txt","rb").read()
```

```python
#shift_jisでデコード
text = binarydata.decode("shift_jis")
```

```python
#「-」5つ並んでいることでテキストをsplitする。今回はリストが3つできる。[0]=作者、題名 [1]=-が5つ並んでいるところで囲まれている行　[2]=本文
text = re.split(r"\-{5,}",text)[2]
```

```python
#最後にあとがきみたいなやつがあるので消す。textにはすでに本文が入っているので今回は[0]
text = re.split(r"底本：",text)[0]
```

```python
#半角スペースの削除
text = text.strip()
```

```python
t=Tokenizer()
results=[]
#改行ごとに分割
lines = text.split("\r\n")
```

```python
#一行ずつ処理を行う
for line in lines:
    s = line
    #|を削除
    s = s.replace('|','')
    #ふりがなを削除
    s = re.sub(r"《.+?》",'',s)
    #説明文を削除
    s = re.sub(r"[#.+?]",'',s)
    s = re.sub(r"［.+?］",'',s)
    #不要なものを削除したので形態素解析を行う
    tokens = t.tokenize(s)
    r = []
    #表層形ごとに処理を行う
    for token in tokens:
        #形態素の活用形が存在しない場合は表層文字（そのままの文字）を格納
        #活用形がある場合は基本形を格納
        #（例）「～である」の「で」は「～だ」の活用形なので「で」の基本形である「だ」として処理を行う
        if token.base_form == "*":
            w = token.surface
        else:
            w = token.base_form
        #品詞を取り出す
        ps = token.part_of_speech
        #part_of_speechには品詞と品詞の細分類3つあり、リスト型となっている。品詞名しか必要ないので[0]で取得
        hinshi = ps.split(',')[0]
        if hinshi in ['名詞','形容詞','動詞','記号']:
            #上記の品詞に該当する場合はリストrに追加していく
            r.append(w)
    #rに対して" " をつけて連結。.stripで前後の空白を除く
    rl = (" ".join(r)).strip()
    results.append(rl)
    #print(rl)
        
```

```python
#モデルファイルに書き出す
#分かち書き→スペースで区切った文字列
wakachigaki_file = "Gingatetsudou.wakachi"
#分かち書きファイルを作成
with open(wakachigaki_file,'w',encoding='utf-8') as fp:
    #結果を改行を含めて書き込みする
    fp.write("\n".join(results))
```

```python
#文ごとに読み込み
data = word2vec.LineSentence(wakachigaki_file)
#Word2Vecのオプション
'''
data→学習に使用するファイル
size→出力するベクトルの次元数
window→一つの単語に対してこの数値分だけ前後をチェックする
hs→学習に階層化ソフトマックスを使用するかどうか(1は使用しない)
min_count→指定した回数より登場回数が少ない単語は破棄する
sg→skip-gramを使用するかどうか(1:使用する0:C-BOWを使用する)
※skip-gramはニューラルネットワークの一種（詳細はggr)
'''
model = word2vec.Word2Vec(data,size=200,window=10,hs=1,min_count=2,sg=1)
#モデルを保存
model.save('Gingatetsudou.model')
print('Complete')
```
